```haskell
---Retorna la longitud de una lista
miLength :: [a] -> Int
miLength [] = 0
miLength (_:xs) = 1 + miLength xs

miLength [2,9,3,4,6,8]
```

```haskell
miLength = foldr (\_ acc -> acc + 1) 0

miLength [2,9,3,4]
```

```haskell
---Retorna el último elemento de una lista
miLast :: [a] -> a
miLast [] = error "No hay un numero"
miLast [x] = x
miLast (_:xs) = miLast xs

miLast [2,6,4]
```

```haskell
miLast = foldl1 (\_ x -> x)

miLast [2,6,5,1,2,3]
```

```haskell
---Duplica cada elemento de una lista de números
duplicar :: [Int] -> [Int]
duplicar [] = []
duplicar (x:xs) = x * 2 : duplicar xs

duplicar [2,3,4]
```

```haskell
duplicar = map (*2)

duplicar [2,3,4]
```

```haskell
---Eliminar los elementos pares de una lista
eliminarPares :: [Int] -> [Int]
eliminarPares [] = []
eliminarPares (x:xs) =
    if x `mod` 2 == 0
    then eliminarPares xs
    else x : eliminarPares xs

eliminarPares [1,2,3,4,5,6,7,8,9]
```

```haskell
eliminarPares = filter odd

eliminarPares [1,2,3,4,5,6,7,8,9]
```
