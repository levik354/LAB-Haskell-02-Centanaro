```haskell
---Divisores de un numero
divisores :: Int -> [Int]
divisores n = [x | x <- [1..n], n `mod` x == 0]

divisores 3
divisores 6
divisores 18
```

```haskell
---Es perfecto
divisores :: Int -> [Int]
divisores n = [x | x <- [1..n-1], n `mod` x == 0]

esPerfecto :: Int -> Bool
esPerfecto n = sum (divisores n) == n

esPerfecto 7
esPerfecto 12
```
