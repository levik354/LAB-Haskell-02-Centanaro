```haskell
--- Celsius a Fahrenheit
procesarLecturas :: (Double -> Double) -> [Double] -> [Double]
procesarLecturas f xs = map f xs
cAf :: Double -> Double
cAf c = c * 1.8 + 32

--Ejemplo:
procesarLecturas cAf [0,12,20]
```

```haskell
--Pares
crearFiltro :: (a -> Bool) -> ([a] -> [a])
crearFiltro f = filter f
soloPares :: [Int] -> [Int]
soloPares = crearFiltro even

--Ejemplo:
soloPares [1..10] == [2,4,6,8,10]
```

```haskell
---La lista dependiendo del rango
enRango :: Int -> Int -> (Int -> Bool)
enRango x y = \a -> a >= x && a <= y
entre5y10 = enRango 5 10

---Ejemplo:
entre5y10 7   -- True
entre5y10 8
entre5y10 11
entre5y10 4
```

```haskell
---Potencia
potenciador :: Int -> (Int -> Int)
potenciador n = \a -> a^n


Ejemplo:

potencia5 = potenciador 5
potencia5 7
```
